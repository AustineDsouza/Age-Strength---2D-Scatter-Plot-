```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset into pandas dataframe
results = pd.read_excel('C:\Austine\Blog\ML\Reference\Age_Strength.xlsx')
```

```python
#Always understand the axis: labels and scale

results.plot(kind='scatter', x='Age', y='Weight');
plt.xlabel('Age')
plt.ylabel('Weights pulled')
plt.title('Weight pulling capacity')
plt.show()

#We can see that the students who are spending more time studies daily have received higher score in maths
```
